```python
#importing all necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  
```

```python
#storing the data as data
data=pd.read_csv('diabetes_model.csv')
```

```python
#checking the initial few columns of the data
data.head()
```

# 1. Create a Count plot of Outcome    

```python
plt.title('Countplot of Outcome')
fig=sns.countplot(x='Outcome',data=data)
#to get the count displayed on the chart
for p in fig.patches:
    fig.annotate(p.get_height() , (p.get_x()+0.35, p.get_height()-150))
```

# 2. Analyze the effect of age on Outcome using Boxplot.

```python
plt.figure(figsize=(8,8))
plt.title('Age v/s Outcome')
sns.boxplot(x='Outcome',y='Age',data=data)
plt.show()
```

Inference: It is observed that chances of being diabetic increases with age. 

# 3. Check for the chances of getting diabetes if it is common in family, using the Daiabetes Pedigree Function.

```python
plt.figure(figsize=(8,8))
sns.boxplot(x='Outcome',y='DiabetesPedigreeFunction',data=data)
plt.show()
```

Inference: It is observed that there is a higher chance of being diabetic if it runs in the family.

# 4. Check for the relationship with Insulin & Glucose and group the datapoints by Outcome.

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x='Insulin',y='Glucose',data=data,hue='Outcome')
plt.show()
```

Inference: There exists a linear relationship between insulin and glucose. 

# 5. Create a heatmap on Correlation of features in the data.

```python
plt.figure(figsize=(12,10)) 
sns.heatmap(data.corr(),annot=True)  
plt.show()
```
